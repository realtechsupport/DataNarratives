```
# HISTOGRAM
# https://mathbitsnotebook.com/Algebra1/StatisticsData/SThistogram.html

import matplotlib.pyplot as plt
import numpy as np

# Use numpy to generate a bunch of random data in a bell curve around 5.
data = 5 + np.random.randn(1000)

xvalues = [m for m in range(len(data))]
plt.plot(xvalues, data, '.')
plt.title("scatter")
plt.show()

plt.bar(data, xvalues, edgecolor='black', linewidth=1.2)
plt.title("bar")
plt.show()

plt.hist(data, bins=20)
plt.title("histogram")
plt.show()

plt.hist(data, bins=20, edgecolor='black', linewidth=1.2)
plt.title("histogram")
plt.show()
```

```
# https://seaborn.pydata.org/tutorial/distributions.html
# colors in seaborn
# https://seaborn.pydata.org/tutorial/color_palettes.html

penguins = pandas.read_csv ('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/penguins.csv')

print(penguins.head())

seaborn.displot(penguins, x=penguins['flipper_length_mm'])
plt.show()
```

```
seaborn.displot(penguins, x=penguins['flipper_length_mm'], stat='probability')
```

```
seaborn.displot(penguins, x=penguins['flipper_length_mm'], hue=penguins['species'])
```

```
seaborn.displot(penguins, x=penguins['flipper_length_mm'], hue=penguins['species'], element='step')
```

```
seaborn.displot(penguins, x=penguins['flipper_length_mm'], hue=penguins['species'], multiple='stack')
```

```
seaborn.displot(penguins, x=penguins['flipper_length_mm'], hue=penguins['species'], kind = 'kde', multiple='stack')
```

```
seaborn.displot(penguins, x=penguins['flipper_length_mm'], hue=penguins['species'], multiple='dodge')
```

```
seaborn.displot(penguins, x=penguins['flipper_length_mm'], col=penguins['species'], multiple='dodge')
```
