```
#PIE CHARTS
import pandas
penguins = pandas.read_csv ('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/penguins.csv')
print(penguins.head())
print(list(penguins))
```

```
import matplotlib.pyplot as plt
import numpy

# color chart
# https://html-color-codes.info/
ave_flipper_length = numpy.mean(penguins['flipper_length_mm'])
ave_bill_length = numpy.mean(penguins['bill_length_mm'])
ave_bill_depth = numpy.mean(penguins['bill_depth_mm'])

mylabels = 'flipper_length', 'bill_length', 'bill_depth'
sections = [ave_flipper_length, ave_bill_length, ave_bill_depth]

colors = ['#ff9999','#66b3ff','#99ff99']
```

```
#plt.pie(sections, labels=labels, colors=colors, autopct = '%1.2f%%')
plt.pie(sections, labels=mylabels) # simplest !!
plt.axis('equal') 
plt.tight_layout()
plt.title('Pie Chart Example')
plt.show()
```
